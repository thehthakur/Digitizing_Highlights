```python
import torch
import segmentation_models_pytorch as smp
from torchvision import transforms
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Number of classes in your dataset (modify this as per your data)
NUM_CLASSES = 2  # Example: 5 classes for multi-class segmentation

# Load the model
model = smp.Unet(
    encoder_name="resnet34",        # choose encoder
    encoder_weights="imagenet",     # use `imagenet` pre-trained weights
    in_channels=3,                  # input channels (e.g. 3 for RGB images)
    classes=NUM_CLASSES             # set number of output classes
)


# Preprocessing (resize and normalize)
preprocess = transforms.Compose([
    transforms.Resize((256, 256)),  # Resize to the input size expected by the model
    transforms.ToTensor(),          # Convert to tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # ImageNet normalization
])

# Load the input image
image = Image.open('bro_and_ghodi.jpg')

# Preprocess the image
input_tensor = preprocess(image)
input_tensor = input_tensor.unsqueeze(0)  # Add batch dimension

# Move model and input to device (CPU or GPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
input_tensor = input_tensor.to(device)

# Set the model to evaluation mode
model.eval()

# Run inference
with torch.no_grad():
    output = model(input_tensor)

# Post-process the output
# output shape: [batch_size, num_classes, height, width]
# Use argmax to get the class with the highest probability for each pixel
output = torch.argmax(output, dim=1)  # Output shape: [batch_size, height, width]

# Convert the tensor to numpy
output_np = output.cpu().numpy().squeeze()  # Remove batch dimension

# Visualize the segmentation result
plt.imshow(output_np, cmap='jet')  # Use 'jet' or another colormap for multiple classes
plt.show()
```
